```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import measure
```

```python
im1 = cv2.imread('screenshot1.png')
im2 = cv2.imread('screenshot2.png')
im1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
im2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)


plt.figure(figsize=(90, 30))
plt.subplot(411),plt.imshow((im1-im2),'gray')
ret,thresh1 = cv2.threshold(im1-im2,5,255,cv2.THRESH_BINARY)
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(thresh1,kernel,iterations = 2)
opening = cv2.morphologyEx(thresh1, cv2.MORPH_OPEN, kernel)
plt.subplot(412),plt.imshow(thresh1,'gray')
plt.subplot(413),plt.imshow(dilation,'gray')
plt.show()

```

```python
#from skimage.morphology import watershed
#segmentation = watershed(elevation_map, markers)
def getroi(image,thd): 
    comp = measure.label(image)
    backtorgb = cv2.cvtColor(image,cv2.COLOR_GRAY2RGB)
    # apply bounding boxes
    for region in measure.regionprops(comp):
        # retrieve x, y, height and width
        y, x, y2, x2 = region.bbox
        height = y2 - y
        width = x2 - x
        if ( width >image.shape[1]*thd or height >image.shape[0]*thd):  
            cv2.rectangle(backtorgb,(x,y),(x2,y2),(0,255,0),3)
    return backtorgb
    
```

```python
image = getroi(dilation,0.1)
```

```python
plt.figure(figsize=(30, 30))
print(image.shape)
plt.imshow(image,'gray')
```
